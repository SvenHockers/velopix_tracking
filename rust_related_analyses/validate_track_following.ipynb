```python
import json 
import os
import sys
sys.path.append("../")

from py_modules.algorithms.parameterised_func.TrackFollowing import track_following as PythonImpl
from py_modules.event_model.event_model import event as ev
from py_modules.validator import validator_lite as vl
from velopix_tracking import TrackFollowing, Event, validate_print
```

```python
events = []
py_events = []
json_events = []
n_files = 2

for i in range(0, n_files):
    if i == 51:
        """
        There's an issue with event 51 -> module_prefix_sum contains value 79 twice resulting in and indexing error when loading the event
        """
        print(f"Skipping problematic file: velo_event_{i}.json")
    else:    
        print(f"Loading file: velo_event_{i}.json")
        event_file = open(os.path.join("../DB/raw", f"velo_event_{i}.json"))
        json_data = json.loads(event_file.read())
        events.append(Event(json_data))
        py_events.append(ev(json_data))
        json_events.append(json_data)
        event_file.close()
```

```python
model = TrackFollowing()
model_p = PythonImpl()
tracks = []
tracks_r = []
for e in py_events:
    tracks.append(model_p.solve(e))
for e in events:
    tracks_r.append(model.solve(e))

tracks_r_p = model.solve_parallel(events)
```

```python
print(tracks[0])
print(tracks_r[0])
print(tracks_r_p[0])
```

```python
validate_print(json_events, tracks_r)
```

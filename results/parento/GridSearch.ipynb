```python
from velopix import validate_print, SearchByTripletTrie, Event
import os 
import sys
import json
```

```python
events = []
py_event = []
n_files = 100

for i in range(0, n_files):
    if i == 51:
        """
        There's an issue with event 51 -> module_prefix_sum contains value 79 twice resulting in and indexing error when loading the event
        """
        print(f"Skipping problematic file: velo_event_{i}.json")
    else:    
        print(f"Loading file: velo_event_{i}.json")
        event_file = open(os.path.join("../../DB/raw", f"velo_event_{i}.json"))
        json_data = json.loads(event_file.read())
        events.append(Event(json_data)) # type: ignore
        py_event.append(json_data)
        event_file.close()
```

```python
scatter = [0.000775]
```

```python
for value in scatter:
    S = SearchByTripletTrie(
        max_scatter=value,
        min_strong_track_length=2,
        allowed_missed_modules=2
    )
    Tracks = S.solve_parallel(events)

    validate_print(py_event, Tracks, False)
```
